## 2. Descriptive Statistics

```python
# Calculate basic statistics for the "VALUE" column

mean_value = rsd['VALUE'].mean()
median_value = rsd['VALUE'].median()
mode_value = rsd['VALUE'].mode()[0]
std_dev_value = rsd['VALUE'].std()

# Print the results

print(f"Mean VALUE: {mean_value}")
print(f"Median VALUE: {median_value}")
print(f"Mode VALUE: {mode_value}")
print(f"Standard Deviation of VALUE: {std_dev_value}")
```

## 3. Time Series Analysis

```python
# Convert 'REF_DATE' to datetime

rsd['REF_DATE'] = pd.to_datetime(rsd['REF_DATE'])
print(rsd.dtypes)
```

```python
# Set 'REF_DATE' as the index

rsd.set_index('REF_DATE', inplace=True)
```

```python
# Resample to monthly frequency and calculate the sum

monthly_sales = rsd['VALUE'].resample('M').sum()

# Plot the resampled data

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, label='Monthly Sales')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Resample the data by quarter for a more manageable plot

quarterly_sales = rsd['VALUE'].resample('Q').sum()

plt.figure(figsize=(12, 5))

# Quarterly Sales Trends

plt.subplot(2, 1, 2)
quarterly_sales.plot(kind='bar', color='skyblue')
plt.title('Quarterly Sales Trends')

plt.tight_layout()
plt.show()
```

```python
# Yearly Sales Trends

quarterly_sales = rsd['VALUE'].resample('Q').sum()

plt.figure(figsize=(12, 5))

plt.subplot(3, 1, 3)
yearly_sales = rsd['VALUE'].resample('Y').sum()
yearly_sales.plot(kind='bar', color='salmon')
plt.title('Yearly Sales Trends')

plt.tight_layout()
plt.show()
```

```python

from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose time series data with frequency 'Q' for quarterly data

result = seasonal_decompose(rsd['VALUE'], model='additive', period=4)

# Adjust figure size and layout

plt.figure(figsize=(12, 8))

# Plot the original time series

plt.subplot(4, 1, 1)
plt.plot(rsd['VALUE'], label='Original Sales')
plt.title('Original Sales')

# Plot the trend component

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.title('Trend Component')

# Plot the seasonal component

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.title('Seasonal Component')

# Plot the residual component

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.title('Residual Component')

# Adjust layout and show the plot

plt.tight_layout()
plt.show()
```

## 4. Customer and Product Analysis

```python
rsd.head()
```

```python
# Analyze customer demographics and purchasing behavior

customer_demographics = rsd.groupby('GEO')['VALUE'].agg(['count', 'mean', 'sum'])

product_analysis = rsd.groupby('Sales')['VALUE'].agg(['count', 'mean', 'sum'])

print("Customer Demographics:")
print(customer_demographics)

print("\nProduct Analysis:")
print(product_analysis)
```

```python
# Calculate the average transaction value by product

average_transaction_by_product = rsd.groupby('Sales')['VALUE'].mean()

# Print the results

print("Average Transaction Value by Product:")
print(average_transaction_by_product)
```

```python
# Calculate total sales for each product category

product_sales = rsd.groupby('Sales')['VALUE'].sum().sort_values(ascending=False)

# Plot

plt.figure(figsize=(10, 8))
sns.barplot(x=product_sales.index, y=product_sales.values, palette='Set2')
plt.title('Purchasing Behavior for Different Products')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the values on top of the bars

for i, value in enumerate(product_sales.values):
    plt.text(i, value, f'{value:,}', ha='center', va='bottom')

plt.show()
```

```python
# Calculate the number of transactions by product category

transaction_count_by_product = rsd.groupby('Sales')['VALUE'].count().sort_values(ascending=False)

# Display the result

print("Number of Transactions by Product:")
print(transaction_count_by_product)
```

```python
# Plot purchasing behavior for different products

plt.figure(figsize=(45, 10))
plt.subplot(1, 2, 1)
sns.barplot(x=product_analysis.index, y='count', data=product_analysis)
plt.title('Number of Transactions by Product')

```

```python
# Calculate average transaction value by product
avg_transaction_by_product = rsd.groupby('Sales')['VALUE'].mean().sort_values(ascending=False)

# Plot the average transaction value by product
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_transaction_by_product.index, y=avg_transaction_by_product.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Average Transaction Value by Product')
plt.xlabel('Product')
plt.ylabel('Average Transaction Value')
plt.show()
```
